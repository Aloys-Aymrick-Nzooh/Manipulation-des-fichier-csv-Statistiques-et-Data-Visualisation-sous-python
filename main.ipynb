Pretraitement des données

```python


def new_fichier(a):
    clients= pd.read_csv(a+".csv")
    clients_new = clients.dropna(how='all')
    clients_new.to_csv(a+"_new.csv", index= False )

#new_fichier("clients")
#new_fichier("emplyes")

###Fonction qui remplace le code postal null par "0000"
def remp_codpos_null(colonne):
    emplyes = pd.read_csv("clients_new.csv")
    emplyes[colonne] = emplyes[colonne].fillna("0000")
    emplyes.to_csv("emplyes_new.csv", index = False )
    
#remp_codpos_null("postalCode")
```

Statisques

```python
#Stats 1
def num_missing(x):
    return sum(x.isnull())
def stats(collone, num_fichier):
    #b = int(input("Entrer 0 pour les clients et 1 pour les employée "))
    if num_fichier == 0 :
        clients = pd.read_csv('clients_new.csv')
    if num_fichier == 1:
        clients=pd.read_csv('emplyes_new.csv')
    #collone = input("Choisir une collone de notre du dataframe:  ")
    if clients[collone].dtype == 'object':
        print(clients[collone].value_counts())
        print(f'Valeurs manquantes : {num_missing(clients[collone])}')
        clients[collone].value_counts().plot(kind='bar')
        plt.show()
    if clients[collone].dtype == 'float64':
        print("la moyenne est: " ,clients[collone].mean())
        print("la min est: ", clients[collone].min())
        print("la max est: ", clients[collone].max())
        print('Valeurs manquantes: ',num_missing(clients[collone]))
        clients[collone].plot.hist(bins=20)
        plt.show()
    
#stats("city",0 )

#stats 2
def nom_emp_inf(num_empl):
    emplyes = pd.read_csv("emplyes_new.csv")
    k = 0
    for i in list(emplyes["reportsTo"]):
        if i == num_empl:
            k = k + 1
    print(f'{num_empl} est supérieur à {k} employees')
    return(k)
    
        
#nom_emp_inf(1002)

#Pour tout les employées
def sup_pour_tout_les_empl():
    emplyes = pd.read_csv("emplyes_new.csv")
    for i in list(emplyes["employeeNumber"]):
        nom_emp_inf(i)
        
#Stats 3
def nbr_client_encharge(a):
    client = pd.read_csv("clients_new.csv")
    k = 0
    for i in list(client["salesRepEmployeeNumber"]):
        if i == a:
            k +=1
    print(f'{a} est en charge de {k} clients')
    return(k)

#nbr_client_encharge(1504)        
 
def client_encharge_pour_tout_empl():
    emplye = pd.read_csv("emplyes_new.csv")
    for i in list(emplye["employeeNumber"]):
        nbr_client_encharge(i)

#client_encharge_pour_tout_empl()
```

Croisement des données

```python
##1
def sum_credit_client_employee(a):
    #emplye = pd.read_csv("emplyes_new.csv")
    client = pd.read_csv("clients_new.csv")
    list_credit_client = list(client["creditLimit"])
    list_empl_resp_client = list(client["salesRepEmployeeNumber"])
    k = 0
    for i in range(0,len(list_empl_resp_client)) :
        if  list_empl_resp_client[i] == a :
            k = k + list_credit_client[i]
            i += 1
    print(f'{a} a un credit total de ses clients = {k}')
    return(k)
           
   #sum_credit_client_employee(1165)    


   ##2
def sum_credit_pour_tout_employee():
    emplye = pd.read_csv("emplyes_new.csv")
    for num_employee in list(emplye["employeeNumber"]):
        sum_credit_client_employee(num_employee)
    fichier = pd.DataFrame(index=list(emplye["employeeNumber"]),
                              columns=["Total Credit"],
                              data=list(emplye["employeeNumber"].apply(sum_credit_client_employee)))
    fichier.to_csv("Total Credit.csv")
#sum_credit_pour_tout_employee()

   ##3
def credit_tot_pour_une_ville(a):
    clients = pd.read_csv("clients_new.csv")
    #villes_unique = list(set(clients["city"]))
    list_credit_client = list(clients["creditLimit"])
    villes = list(clients["city"])
    k = 0
    for i in range(0,len(villes)):
        if villes[i] == a:
            k = k + list_credit_client[i]
            #print(villes.index(vil))
            i += 1
    print(f'{a}: montant credit client basée: {k}')
    return(k)
           
#credit_tot_pour_une_ville("NYC")   

def credit_tot_ville(): 
    clients = pd.read_csv("clients_new.csv")
    villes_unique = list(set(clients["city"]))
    city= pd.DataFrame(data = villes_unique,
                          columns=["city"])
    for ville_uni in villes_unique:
        credit_tot_pour_une_ville(ville_uni)
        fichier = pd.DataFrame(index=list(city["city"]),
                              columns=["Total Credit"],
                              data=list(city["city"].apply(credit_tot_pour_une_ville)))
    fichier.to_csv("Total Credit Par Ville.csv")
       
#credit_tot_ville()

def credit_tot_un_pay(a):
    clients = pd.read_csv("clients_new.csv")
    #villes_unique = list(set(clients["city"]))
    list_credit_client = list(clients["creditLimit"])
    pays = list(clients["country"])
    k = 0
    for i in range(0,len(pays)):
        if pays[i] == a:
            k = k + list_credit_client[i]
            i += 1
    print(f'{a}: montant credit client basée: {k}')
    return(k)

   #credit_tot_un_pay("France")
       
def credit_tot_pay():
    clients = pd.read_csv("clients_new.csv")
    pays_unique = list(set(clients["country"]))
    pays= pd.DataFrame(data = pays_unique,
                          columns=["country"])
    for pay_uni in pays_unique:
        credit_tot_un_pay(pay_uni)
        fichier = pd.DataFrame(index=list(pays["country"]),
                              columns=["Total Credit"],
                              data=list(pays["country"].apply(credit_tot_un_pay)))
    fichier.to_csv("Total Credit Par Pays.csv")
       
   #credit_tot_pay()
```
